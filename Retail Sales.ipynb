```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
data.head(2)
```

```python
data["Store Location"]=le.fit_transform(data["Store Location"])
data["Product Category"]=le.fit_transform(data["Product Category"])
data["Day of the Week"]=le.fit_transform(data["Day of the Week"])
data["Holiday Effect"]=le.fit_transform(data["Holiday Effect"])
```

```python
data.head(5)
```

```python
plt.figure(figsize=(9,7))
sns.heatmap(data.corr(),annot=True)
```

```python
data=data.drop(["Store Location"],axis=1)
```

```python
data.head(5)
```

```python
data['timestamp'] = data['Date'].astype(int) / 1e9  
data.drop('Date', axis=1, inplace=True) 
```

```python
x=data.drop(["Sales Revenue (USD)"],axis=1)
y=data["Sales Revenue (USD)"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
rfr=RandomForestRegressor(n_estimators=10)
rfr.fit(x_train,y_train)
```

```python
y_pred=rfr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_pred,y_test)
print(r2*100)
```
